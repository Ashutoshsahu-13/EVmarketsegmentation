```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
df1=df.iloc[:,:12]
df1
```

```python
pca_data=preprocessing.scale(df1)
pca=PCA(n_components=12)
pc = pca.fit_transform(df1)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11','pc12']
pf = pd.DataFrame(data=pc,columns=names)
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
# correlation between the original variables and principal components
loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = df1.columns.values
loadings_df = loadings_df.set_index('variable')
```

```python
# Heat map for original variables and principal components
plt.rcParams['figure.figsize'] = (25,25)
ax = sns.heatmap(loadings_df, annot=True)
plt.show()
```

```python
plt.figure(figsize=(15,14))
from sklearn.cluster import KMeans
# Get the inertia values for number of clusters varying from 2 to 10
for j in range(2,11):
    inertia = []
    for i in range(2, 11):
        kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
        kmeans.fit(df1)
        inertia.append(kmeans.inertia_)

    # Plot the intertia values for these clusters to choose the appropriate number of clusters
    plt.subplot(3,4,j)
    plt.plot(range(2, 11), inertia,marker = "8")
    plt.title('The Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
plt.show()
```

```python
kmeans3 = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
kmeans3.fit(df1)
df1['cluster'] = kmeans3.labels_
print(kmeans3.cluster_centers_)
```

```python
from bioinfokit.visuz import cluster
cluster.screeplot(obj=[pc_list, pca.explained_variance_ratio_],show=True,dim=(10,5))
```

```python
pca_scores = PCA().fit_transform(df1)

# get 2D biplot
cluster.biplot(cscore=pca_scores, loadings=loadings, labels=df1.columns.values, var1=round(pca.explained_variance_ratio_[0]*100, 2),
    var2=round(pca.explained_variance_ratio_[1]*100, 2),show=True,dim=(10,5))
```

```python
marry = df1.groupby('cluster')['Marrital Status'].mean()
marry = marry.to_frame().reset_index()
marry
```

```python
df['Car Type'] = df['Car Type'].replace({'Hatchback':1,'SUV':3,'Luxuray':4})
df['Car Type'] = df['Car Type'].replace({'sedan':2})
```

```python
df['Car Type'] = df['Car Type'].astype('int')
ct = df.groupby('cluster')['Car Type'].mean()
ct = ct.to_frame()
ct
```

```python
edu = df1.groupby('cluster')['Education'].mean()
edu = edu.to_frame()
edu
```

```python
dep = df1.groupby('cluster')['No of Dependents'].mean()
dep = dep.to_frame()
dep
```

```python
ts = df1.groupby('cluster')['Total Salary'].mean().to_frame()
ts
```

```python
car_price = df.groupby('cluster')['Price'].mean().to_frame()
car_price
```

```python
segment = Fc.merge(wife_salary, on='cluster', how='left').merge(salary, on='cluster', how='left').merge(ts,on = 'cluster',how = 'left')
segment = segment.merge(age, on='cluster', how='left').merge(marry, on='cluster', how='left').merge(prof,on = "cluster",how = "left").merge(edu,on = 'cluster',how = 'left').merge(dep,on = 'cluster',how = 'left').merge(ct,on = 'cluster',how = 'left').merge(car_price,on = 'cluster',how = 'left')
```
